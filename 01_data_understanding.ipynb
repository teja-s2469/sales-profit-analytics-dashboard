```python
df.duplicated().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df.info()
```

```python
# EDA

total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print("Total Sales: ", total_sales)
print("Total Profit: ", total_profit)
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Month Name'] = df['Order Date'].dt.strftime('%B')


```

```python
monthly_sales = df.groupby('Order Month')['Sales'].sum()
monthly_sales
df.groupby('Month Name')['Sales'].sum()
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

## Monthly Sales Trend Analysis

### Objective
To analyze how sales fluctuate across different months and identify seasonal patterns.

### Observation
- Sales show noticeable variation from month to month.
- Certain months consistently record higher sales compared to others.
- Peak sales periods indicate seasonal or festive demand trends.

### Business Insight
- Businesses can plan marketing campaigns and inventory management around peak sales months.
- Low-performing months may require promotional offers or targeted advertising.
- Understanding monthly trends helps organizations forecast demand, optimize stock levels, and improve revenue planning.


```python
# region wise profit
region_profit = df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
region_profit
```

```python
region_profit.plot(kind='bar', title='Region Wise Profit')
```

## Region Wise Profit Analysis

### Objective

### Observation

### Business Insight
